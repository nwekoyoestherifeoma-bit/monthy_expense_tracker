```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def welcome_message():
    print("=      EXPENSE TRACKER")
    print("="  * 50)
```

```python
def add_expense():
    while True:
         expense = input("Enter expense name: ")
         amount = float(input("enter expense amount: $"))

         expenses.append(expense)
         amounts.append(amount)

         choice = input("do you want to add another expense? (yes/no:) ").lower()

         if choice == "no":
             break
```

```python
def create_report():
    df = pd.DataFrame({
        "Expense": expenses,
        "Amount": amounts
    })

    expense_array = np.array (amounts)

    total = np.sum(expense_array)
    balance = monthly_earnings - total
    average = np.mean(expense_array)
    highest = np.max(expense_array)
    lowest = np.min(expense_array)

    print("\nExpense Report")
    print(df)

    print(f"\nTotal Expenses: ${total:,.2f}")
    print(f"Average Expense: ${average:,.2f}")
    print(f"Highest Expense: ${highest:,.2f}")
    print(f"Lowest Expense: ${lowest:,.2f}")
    print(f"Monthly Earnings: ${monthly_earnings:,.2f}")
    print(f"Remaining Balance: ${balance:,.2f}")




```

```python
def show_graph():
    plt.figure(figsize=(6, 3))

    plt.bar(expenses, amounts)

    plt.title("Expense Tracker")
    plt.xlabel("Expenses")
    plt.ylabel("Amount ($)")

    plt.show()

```
